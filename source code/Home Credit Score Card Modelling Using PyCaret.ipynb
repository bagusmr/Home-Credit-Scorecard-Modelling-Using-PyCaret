# Data Preprocessing

## Feature Engineering

```python
# Replacing CODE GENDER values
train['CODE_GENDER'].replace (to_replace = 'XNA', value = train['CODE_GENDER'].mode()[0], inplace = True)

# Replacing NAME_FAMILY_STATUS values 
single = ['Single / not married','Separated','Widow']
married = ['Civil marriage']

train['NAME_FAMILY_STATUS'].replace(to_replace = single, value = 'Single', inplace = True)
train['NAME_FAMILY_STATUS'].replace(to_replace = married, value = 'Married', inplace = True)
train['NAME_FAMILY_STATUS'].replace(to_replace = 'Unknown', value = train['NAME_FAMILY_STATUS'].mode()[0], inplace = True)
```

## Handling Missing Values

```python
# Fill as if nan values means the expert did not give the score, equals 0
train['EXT_SOURCE_2'].fillna(0, inplace = True)
train['EXT_SOURCE_3'].fillna(0, inplace = True)
```

## Feature Extraction

```python
# Feature extraction
train['YEAR_BIRTH'] = round((train['DAYS_BIRTH'].abs()/365),0)
train['YEAR_EMPLOYED'] = round((train['DAYS_EMPLOYED'].abs()/365),0)
train['YEAR_REGISTRATION'] = round((train['DAYS_REGISTRATION'].abs()/365),0)
train['YEAR_ID_PUBLISH'] = round((train['DAYS_ID_PUBLISH'].abs()/365),0)
train['YEAR_LAST_PHONE_CHANGE'] = round((train['DAYS_LAST_PHONE_CHANGE'].abs()/365),0)
train['AVG_EXT_SOURCE'] = (train['EXT_SOURCE_2'] + train['EXT_SOURCE_3'])/2
train['AVG_FLAG_DOCS'] = (train['FLAG_DOCUMENT_2'] + train['FLAG_DOCUMENT_3'] + train['FLAG_DOCUMENT_4'] + train['FLAG_DOCUMENT_5'] + train['FLAG_DOCUMENT_6'] +
                          train['FLAG_DOCUMENT_7'] + train['FLAG_DOCUMENT_8'] + train['FLAG_DOCUMENT_9'] + train['FLAG_DOCUMENT_10'] + train['FLAG_DOCUMENT_11'] +
                          train['FLAG_DOCUMENT_12'] + train['FLAG_DOCUMENT_13'] + train['FLAG_DOCUMENT_14'] + train['FLAG_DOCUMENT_15'] + train['FLAG_DOCUMENT_16'] +
                          train['FLAG_DOCUMENT_17'] + train['FLAG_DOCUMENT_18'] + train['FLAG_DOCUMENT_19'] + train['FLAG_DOCUMENT_20'] + train['FLAG_DOCUMENT_21'])/20
```

## Feature Selection

```python
train.drop(columns=['EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH','DAYS_LAST_PHONE_CHANGE',
                    'FLAG_DOCUMENT_2','FLAG_DOCUMENT_3','FLAG_DOCUMENT_4','FLAG_DOCUMENT_5','FLAG_DOCUMENT_6','FLAG_DOCUMENT_7', 
                    'FLAG_DOCUMENT_8','FLAG_DOCUMENT_9','FLAG_DOCUMENT_10','FLAG_DOCUMENT_11','FLAG_DOCUMENT_12','FLAG_DOCUMENT_13', 
                    'FLAG_DOCUMENT_14','FLAG_DOCUMENT_15','FLAG_DOCUMENT_16','FLAG_DOCUMENT_17','FLAG_DOCUMENT_18','FLAG_DOCUMENT_19', 
                    'FLAG_DOCUMENT_20','FLAG_DOCUMENT_21','FLAG_MOBIL','FLAG_EMP_PHONE','FLAG_WORK_PHONE','FLAG_CONT_MOBILE',
                    'FLAG_PHONE','FLAG_EMAIL','REGION_RATING_CLIENT_W_CITY','HOUR_APPR_PROCESS_START','REG_REGION_NOT_LIVE_REGION',
                    'REG_REGION_NOT_WORK_REGION','LIVE_REGION_NOT_WORK_REGION','REG_CITY_NOT_LIVE_CITY','REG_CITY_NOT_WORK_CITY',
                    'LIVE_CITY_NOT_WORK_CITY'], inplace = True)
```
